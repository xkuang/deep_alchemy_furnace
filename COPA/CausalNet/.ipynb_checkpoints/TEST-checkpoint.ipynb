# Causal Word2vec

## Import Libraries

```python
from keras.layers import Input, Embedding, Dense, Flatten, BatchNormalization, Activation
from keras.models import Model
import keras.backend as K
from keras.callbacks import*
from tqdm import tqdm
import numpy as np
from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

## Set Hyperparameters

```python
VOCAB_SIZE = len(word2index)+1
EMBEDDING_DIM = 256
BATCH_SIZE = 1024
NUM_EPOCHS = 256
STEPS_PER_EPOCH = 2048
```

## Build Graph

```python
def build():
    """
    Build embedding matrix
    """
    K.clear_session()
    seq = Input(shape=(None,), dtype='int64')
    emb = Embedding(VOCAB_SIZE,
                    EMBEDDING_DIM,
                    mask_zero=False,
                    trainable=True)(seq)
    model = Model(inputs=seq, outputs=emb)
    return model
```

```python
K.clear_session()

embedding = build()

inputs = Input(shape=(1,), name='INPUT', dtype='int64')
embSeq = embedding(inputs)
embSeq = Flatten(name='FLATTEN')(embSeq)
dense = Dense(VOCAB_SIZE, activation=None, name='DENSE')(embSeq)
bn = BatchNormalization(name='BN')(dense)
out = Activation('softmax', name='OUTPUT')(bn)
model = Model(inputs=inputs, outputs=out)
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam')
```

```python
def clean_data_1(x, y):
    """
    Delete redundant data
    """
    A = x.copy()
    B = y.copy()
    for i in range(500):
        if any(A[i] != B[i]):
            for a in range(MAX_LEN):
                for b in range(MAX_LEN):
                    if A[i][a] == B[i][b]:
                        A[i][a] = 0
                        B[i][b] = 0
    return (A, B)

def clean_data_2(d):
    """
    Delete redundant 0
    """
    d = [[i for i in s if i != 0] for s in d]
    d = [[0] if s == [] else s for s in d] 
    return d
```

```python
def causal_strength_cos(a1, a2):
    """
    Compute causal strength with cosine similarity
    """
    Score1 = []
    for i in range(len(a1)):
        a1Len = len(a1[i])
        a2Len = len(a2[i])
        S1 = []
        for a in a1[i]:
            a = normed_embedding[a]
            for b in a2[i]:
                b = normed_embedding[b]
                S1.append(np.abs(np.dot(a, b)))
        Score1.append(sum(S1) / (a1Len + a2Len))
    return Score1
```

```python
def causal_strength(a1, a2):
    """
    Compute causal strength
    """
    Score1 = []
    for i in range(len(a1)):
        a1Len = len(a1[i])
        a2Len = len(a2[i])
        result1 = np.zeros((a1Len, VOCAB_SIZE))
        S1 = []
        for j in range(a1Len):
            result1[j, :] = model.predict(np.array([a1[i][j]]))
            for k in a2[i]:
                S1.append(np.log(result1[j, k]))
        Score1.append(sum(S1) / (a1Len + a2Len))
    return Score1
```

```python
def accuracy(metric, a1, a2, b1, b2, actu, show=True):
    """
    Calculate Accuracy
    """
    A1 = metric(a1, a2)
    A2 = metric(b1, b2)
    pred = []
    for i in range(len(actu)):
        if A1[i] > A2[i]:
            pred.append(1)
        else:
            pred.append(2)
    S = sum([1 for i in range(len(pred)) if pred[i] == actu[i]])
    ACC = S / len(actu)
    if show:
        print('Accuracy: \t%.9f' % (ACC))
    #return np.array([ACC])
    return pred
```

# GRU Classification 

## Load Embedding

```python
filename = '/Users/lizhn7/Downloads/EXPERIMENT/COPA/CausalNet/word2vec/cp_logs/weights.241-6.465599.hdf5'
model.load_weights(filename)

for layer in embedding.layers:
    layer.trainable=False
```

## Set Hyperparameters

```python
yTrain = np.array([i if i == 1 else 0 for i in yTrain])
```

```python
DELTA = 0.009
MAX_LEN = 12
GRU_SIZE = 64
DROPOUT_RATE = 0.5
BATCH_SIZE = 200
NUM_EPOCHS = 256
L2_NORM = 1e-4
WEIGHT_CONSTRAINT = 3.
```

## Import Libraries

```python
from keras.layers import Input, Masking, Dropout, GRU, Dense, BatchNormalization, Activation
from keras.models import Model
import keras.backend as K
from keras.regularizers import l2
from keras.constraints import maxnorm
from keras.callbacks import*
```

## Build Graph

```python
def hinge(y_true, y_pred):
    """
    Define hinge loss function
    """
    return K.mean(K.maximum(DELTA - y_true * y_pred, 0.), axis=-1)
```

```python
def GRU_Classification():
    """
    Build GRU Classification model
    """
    inputs = Input(shape=(1,), name='INPUT')
    emb_seq = embedding(inputs)
    #emb_seq = Masking(mask_value=0., name='MASKING')(emb_seq)
    gru = GRU(GRU_SIZE, 
              return_sequences=True, 
              implementation=0, 
              dropout=DROPOUT_RATE, 
              recurrent_dropout=DROPOUT_RATE,
              kernel_regularizer=l2(L2_NORM),
              recurrent_regularizer=l2(L2_NORM),
              name='GRU')(emb_seq)
    gru = GlobalAveragePooling1D()(gru)
    gru = Dropout(DROPOUT_RATE, name='DROPOUT')(gru)
    dense = Dense(1, activation=None, name='DENSE')(gru)
    bn = BatchNormalization(name='BN')(dense)
    outputs = Activation('sigmoid', name='SIGMOID')(bn)
    model = Model(inputs=[inputs], outputs=[outputs])
    model.compile(loss='binary_crossentropy', optimizer='rmsprop')
    return model
```

```python
GRU_C = GRU_Classification()
```

## Train

```python
from tqdm import tqdm
import numpy as np
```

```python
def accuracy(model, a1, a2, actu, show=True):
    """
    Calculate Accuracy
    """
    A1 = model.predict(a1)
    A2 = model.predict(a2)
    pred = []
    for i in range(len(A1)):
        if A1[i] > A2[i]:
            pred.append(1)
        else:
            pred.append(2)
    S = sum([1 for i in range(len(pred)) if pred[i] == actu[i]])
    ACC = S / len(actu)
    if show:
        print('Accuracy: \t%.9f' % (ACC))
    return np.array([ACC])

def plot_acc(acc, num_epochs):
    """
    Plot Accuracy
    """
    print('MAX Accuracy: \t%.3f' % (max(acc)))
    epochs = list(range(1, num_epochs+1))
    plt.figure(figsize=(10, 8))
    plt.plot(epochs, acc, label="Accuracy", color="red", linewidth=1)
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.xticks([i for i in range(1, len(acc), len(acc)//10)])
    plt.grid(True)  
    plt.legend()
    plt.show()
    
def plot_loss(log):
    """
    Plot Loss
    """
    plt.figure(figsize=(10, 8))
    loss = log['loss']
    if 'val_loss' in log:
        val_loss = log['val_loss']
        plt.plot(val_loss, color="r", label="Val Loss")
    plt.plot(loss, color="g", label="Train Loss")
    plt.ylabel("Loss")
    plt.xlabel("Epochs")
    plt.legend(loc="best")
    plt.show()
```

```python
GRU_C = GRU_Classification()
TEST_ACC = []
VAL_ACC = []
history = {}
for e in tqdm(range(NUM_EPOCHS)):
    #print('EPOCHS', e+1)
    t = GRU_C.fit(xTrain, 
                  yTrain,
                  batch_size=BATCH_SIZE,
                  epochs=1,
                  shuffle=True,
                  verbose=0)
    for i, j in t.history.items():
        history[i] = history.get(i, []) + j
    TEST_ACC.append(accuracy(GRU_C, test1, test2, yTest, show=False))
    VAL_ACC.append(accuracy(GRU_C, val1, val2, yVal, show=False))
    
print("Training Loss")
plot_loss(history)
print("Val Accuracy")
plot_acc(VAL_ACC, NUM_EPOCHS)
print("Test Accuracy")
plot_acc(TEST_ACC, NUM_EPOCHS)
```

```python
MAX_LEN = 12
EMBEDDING_DIM = 300
DROPOUT_RATE = 0.5
BATCH_SIZE = 100
HIDDEN_SIZE = 128
NUM_EPOCHS = 2048
```

```python
from keras.layers import Input, Embedding, GlobalAveragePooling1D, Dropout, Dense, BatchNormalization, Activation
from keras import optimizers
from keras.models import Model
import keras.backend as K
from keras.callbacks import*
```

```python
def FastText():
    """
    Build FastText model
    """
    inputs = Input(shape=(MAX_LEN,), name='INPUT')
    emb_seq = Embedding(VOCAB_SIZE, 
                        EMBEDDING_DIM, 
                        weights=[copa_embedding], 
                        mask_zero=False, 
                        input_length=MAX_LEN, 
                        trainable=False, 
                        name='EMBEDDING')(inputs)
    pool = GlobalAveragePooling1D(name='POOLING')(emb_seq)
    pool = Dropout(DROPOUT_RATE, name='DROPOUT_1')(pool)
    dense = Dense(HIDDEN_SIZE, activation=None, name='DENSE_1')(pool)
    bn = BatchNormalization(name='BN_1')(dense)
    act = Activation('relu', name='RELU')(bn)
    act = Dropout(DROPOUT_RATE, name='DROPOUT_2')(act)
    dense2 = Dense(1, activation=None, name='DENSE_2')(act)
    bn2 = BatchNormalization(name='BN_2')(dense2)
    outputs = Activation('tanh', name='tanh')(bn2)
    model = Model(inputs=[inputs], outputs=[outputs])
    sgd = optimizers.SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
    model.compile(loss='hinge', optimizer=sgd)
    return model
```

```python
from keras.layers import Input, Embedding, GlobalAveragePooling1D, concatenate, Dense, BatchNormalization, Activation
from keras.models import Model
from keras import optimizers
import keras.backend as K
from keras.callbacks import*
```

```python
def hinge(y_true, y_pred):
    """
    Define hinge loss function
    """
    return K.mean(K.maximum(DELTA - y_true * y_pred, 0.), axis=-1)
```

```python
def FastText():
    """
    Build FastText model
    """
    inputs = Input(shape=(MAX_LEN,), name='INPUT')
    glove_emb_seq = Embedding(VOCAB_SIZE, 
                        EMBEDDING_DIM, 
                        weights=[copa_embedding], 
                        mask_zero=False, 
                        input_length=MAX_LEN, 
                        trainable=False, 
                        name='GLOVE_EMBEDDING')(inputs)
    causal_emb_seq = embedding(inputs)
    glove_pool = GlobalAveragePooling1D(name='GLOVE_POOLING')(glove_emb_seq)
    causal_pool = GlobalAveragePooling1D(name='CAUSAL_POOLING')(causal_emb_seq)
    pool = concatenate([glove_pool, causal_pool], name='CONCATENATE')
    pool = Dropout(DROPOUT_RATE, name='DROPOUT_1')(pool)
    dense = Dense(HIDDEN_SIZE, activation=None, name='DENSE_1')(pool)
    bn = BatchNormalization(name='BN_1')(pool)
    act = Activation('relu', name='RELU')(bn)
    act = Dropout(DROPOUT_RATE, name='DROPOUT_2')(act)
    dense2 = Dense(1, activation=None, name='DENSE_2')(act)
    bn2 = BatchNormalization(name='BN_2')(dense2)
    outputs = Activation('tanh', name='tanh')(bn2)
    model = Model(inputs=[inputs], outputs=[outputs])
    adam = optimizers.Adam(lr=0.001)
    model.compile(loss=hinge, optimizer=adam)
    return model
```

```python
fasttext = FastText()
```

```python
def accuracy(model, a1, a2, actu, show=True):
    """
    Calculate Accuracy
    """
    A1 = model.predict(a1)
    A2 = model.predict(a2)
    pred = []
    for i in range(len(A1)):
        if A1[i] > A2[i]:
            pred.append(1)
        else:
            pred.append(2)
    S = sum([1 for i in range(len(pred)) if pred[i] == actu[i]])
    ACC = S / len(actu)
    if show:
        print('Accuracy: \t%.9f' % (ACC))
    return np.array([ACC])

def plot_acc(acc, num_epochs):
    """
    Plot Accuracy
    """
    print('MAX Accuracy: \t%.3f' % (max(acc)))
    epochs = list(range(1, num_epochs+1))
    plt.figure(figsize=(10, 8))
    plt.plot(epochs, acc, label="Accuracy", color="red", linewidth=1)
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.xticks([i for i in range(1, len(acc), len(acc)//10)])
    plt.grid(True)  
    plt.legend()
    plt.show()
    
def plot_loss(log):
    """
    Plot Loss
    """
    plt.figure(figsize=(10, 8))
    loss = log['loss']
    if 'val_loss' in log:
        val_loss = log['val_loss']
        plt.plot(val_loss, color="r", label="Val Loss")
    plt.plot(loss, color="g", label="Train Loss")
    plt.ylabel("Loss")
    plt.xlabel("Epochs")
    plt.legend(loc="best")
    plt.show()
```

```python
from keras.layers import Input, Embedding, Dense, Flatten, BatchNormalization, Activation
from keras.models import Model
import keras.backend as K
from keras.callbacks import*
from tqdm import tqdm
import numpy as np
```

```python
VOCAB_SIZE = len(word2index)+1
EMBEDDING_DIM = 256
HIDDEN_SIZE = 512
BATCH_SIZE = 1024
NUM_EPOCHS = 256
STEPS_PER_EPOCH = 2048
```

```python
def build():
    """
    Build embedding matrix
    """
    K.clear_session()
    seq = Input(shape=(None,), dtype='int64')
    emb = Embedding(VOCAB_SIZE,
                    EMBEDDING_DIM,
                    mask_zero=False,
                    trainable=True)(seq)
    model = Model(inputs=seq, outputs=emb)
    return model
```

```python
embedding = build()
```

```python
inputs = Input(shape=(1,), name='INPUT', dtype='int64')
embSeq = embedding(inputs)
embSeq = Flatten(name='FLATTEN')(embSeq)
dense1 = Dense(HIDDEN_SIZE, activation=None, name='DENSE_1')(embSeq)
bn1 = BatchNormalization(name='BN_1')(dense1)
act1 = Activation('relu', name='RELU_1')(bn1)
dense2 = Dense(HIDDEN_SIZE, activation=None, name='DENSE_2')(act1)
bn2 = BatchNormalization(name='BN_2')(dense2)
act2 = Activation('relu', name='RELU_2')(bn2)
dense3 = Dense(VOCAB_SIZE, activation=None, name='DENSE_3')(act2)
bn3 = BatchNormalization(name='BN_3')(dense3)
out = Activation('softmax', name='OUTPUT')(bn3)
model = Model(inputs=inputs, outputs=out)
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam')
```

```python
def causal_strength(a1, a2):
    """
    Compute causal strength
    """
    Score1 = []
    for i in range(len(a1)):
        a1Len = len(a1[i])
        a2Len = len(a2[i])
        result1 = np.zeros((a1Len, VOCAB_SIZE))
        S1 = []
        for j in range(a1Len):
            result1[j, :] = model.predict(np.array([a1[i][j]]))
            for k in a2[i]:
                S1.append(np.log(result1[j, k]))
        Score1.append(sum(S1) / (a1Len + a2Len))
    return Score1
```

```python
def accuracy(a1, a2, b1, b2, actu, show=True):
    """
    Calculate Accuracy
    """
    A1 = causal_strength(a1, a2)
    A2 = causal_strength(b1, b2)
    pred = []
    for i in range(len(actu)):
        if A1[i] > A2[i]:
            pred.append(1)
        else:
            pred.append(2)
    S = sum([1 for i in range(len(pred)) if pred[i] == actu[i]])
    ACC = S / len(actu)
    if show:
        print('Accuracy: \t%.9f' % (ACC))
    #return np.array([ACC])
```

```python
def clean_data_1(x, y):
    """
    Delete redundant data
    """
    A = x.copy()
    B = y.copy()
    for i in range(500):
        if any(A[i] != B[i]):
            for a in range(MAX_LEN):
                for b in range(MAX_LEN):
                    if A[i][a] == B[i][b]:
                        A[i][a] = 0
                        B[i][b] = 0
    return (A, B)

def clean_data_2(d):
    """
    Delete redundant 0
    """
    d = [[i for i in s if i != 0] for s in d]
    d = [[0] if s == [] else s for s in d] 
    return d
```

```python
def causal_strength_cos_pro(a1, a2):
    """
    Compute causal strength with cosine similarity
    """
    Score1 = []
    for i in range(len(a1)):
        a1Len = len(a1[i])
        a2Len = len(a2[i])
        S1 = []
        a = sum(normed_embedding[a1[i]])/(a1Len)
        b = sum(normed_embedding[a2[i]])/(a2Len)
        S1.append(sum([np.abs(np.dot(normed_embedding[m], b)) for m in a1[i]])/(a1Len))
        S1.append(sum([np.abs(np.dot(a, normed_embedding[n])) for n in a2[i]])/(a2Len))
        #for b in a2[i]:
        #    b = normed_embedding[b]
        #    S1.append(np.abs(np.dot(a, b)))
        #Score1.append(S1)
        Score1.append(sum(S1))
    return Score1

def causal_strength_cos(a1, a2):
    """
    Compute causal strength with cosine similarity
    """
    Score1 = []
    for i in range(len(a1)):
        a1Len = len(a1[i])
        a2Len = len(a2[i])
        S1 = []
        for a in a1[i]:
            a = normed_embedding[a]
            for b in a2[i]:
                b = normed_embedding[b]
                S1.append(np.abs(np.dot(a, b)))
        Score1.append(sum(S1) / (a1Len + a2Len))
    return Score1

def accuracy(metric, a1, a2, b1, b2, actu, show=True):
    """
    Calculate Accuracy
    """
    A1 = metric(a1, a2)
    A2 = metric(b1, b2)
    pred = []
    for i in range(len(actu)):
        if A1[i] > A2[i]:
            pred.append(1)
        else:
            pred.append(2)
    S = sum([1 for i in range(len(pred)) if pred[i] == actu[i]])
    ACC = S / len(actu)
    if show:
        print('Accuracy: \t%.9f' % (ACC))
    #return np.array([ACC])
    return pred
```
