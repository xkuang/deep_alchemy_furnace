# Import Libraries

```python
from keras.layers import Input, Embedding, Conv1D, Activation, GlobalMaxPooling1D, Concatenate, Dropout, Dense
from keras.layers.normalization import BatchNormalization
from keras.models import Model
import keras.backend as K
from keras.callbacks import*
from keras import optimizers
from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

# Build Graph

```python
def MWCNN(margin=0.6, filter_sizes=(3, 4, 5), num_filter=100, dropout_rate=0.5):
    """
    Hyperparameters
    ----------
    :param filter_sizes: sizes of convolution filter
    :param num_filter: number of convolution layer filters
    :param dropout_rate: dropout layer rate
    :param margin: loss function coefficient(margin)
    
    Returns
    ----------
    Multiwindow cnn model.
    """
    theta = lambda t: (K.sign(t)+1.)/2.
    loss = lambda y_true, y_pred: -(1-theta(y_true-margin)*theta(y_pred-margin)\
                                    -theta(1-margin-y_true)*theta(1-margin-y_pred))\
                                    *(y_true*K.log(y_pred)+(1-y_true)*K.log(1-y_pred))
    seq = Input(shape=(xTrain.shape[-1],), name='INPUT') 
    emb_seq = Embedding(len(word2index)+1, 
                        embedding.shape[-1], 
                        weights=[embedding], 
                        mask_zero=False, 
                        input_length=xTrain.shape[-1], 
                        trainable=False, 
                        name='EMBEDDING')(seq)
    convs = []
    for f in filter_sizes:
        conv = Conv1D(filters=num_filter,
                      kernel_size=f,
                      padding='same',
                      activation=None,
                      name='CONV_FW_%d' % f)(emb_seq)
        bn = BatchNormalization(name='BN_FW_%d' % f)(conv)
        act = Activation('relu', name='RELU_FW_%d' % f)(bn)
        pool = GlobalMaxPooling1D(name='MAXPOOLING_FW_%d' % f)(act)
        convs.append(pool)
    if len(filter_sizes) > 1:
        out = Concatenate(axis=1, name='CONCATENATE')(convs)
        out = Dropout(dropout_rate, name='DROPOUT')(out)
    else:
        out = Dropout(dropout_rate, name='DROPOUT')(pool)
    #out = Concatenate(axis=1, name='CONCATENATE')(convs)
    #out = Dropout(dropout_rate, name='DROPOUT')(out)
    predict = Dense(1, 
                    activation='sigmoid', 
                    name='OUTPUT')(out)
    model = Model(inputs=seq, outputs=predict)
    model.compile(loss=loss, optimizer='adam')
    return model
```

# Train

```python
from tqdm import tqdm
import numpy as np
```

```python
def data_generator(data, label, batch_size):
    """
    Yield batches of all data
    """
    count = 0
    while True:
        if count >= len(data): 
            count = 0
        x = np.zeros((batch_size, xTrain.shape[-1]))
        y = np.zeros(batch_size)
        for i in range(batch_size):
            n = i + count
            if n > len(data)-1:
                break
            x[i, :] = data[n]
            y[i] = label[n]
        count += batch_size
        yield (x, y)

def accuracy(model, a1, a2, actu, show=True):
    """
    Calculate Accuracy
    """
    A1 = model.predict(a1)
    A2 = model.predict(a2)
    pred = []
    for i in range(len(A1)):
        if A1[i] > A2[i]:
            pred.append(1)
        else:
            pred.append(2)
    S = sum([1 for i in range(len(pred)) if pred[i] == actu[i]])
    ACC = S / len(actu)
    if show:
        print('Accuracy: \t%.9f' % (ACC))
    return np.array([ACC])

def plot_acc(acc, num_epochs):
    """
    Plot Accuracy
    """
    print('MAX Accuracy: \t%.3f' % (max(acc)))
    epochs = list(range(1, num_epochs+1))
    plt.figure(figsize=(10, 8))
    plt.plot(epochs, acc, label="Accuracy", color="red", linewidth=1)
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.xticks([i for i in range(1, len(acc), len(acc)//10)])
    plt.grid(True)  
    plt.legend()
    plt.show()
    
def plot_loss(log):
    """
    Plot Loss
    """
    plt.figure(figsize=(10, 8))
    loss = log['loss']
    if 'val_loss' in log:
        val_loss = log['val_loss']
        plt.plot(val_loss, color="r", label="Val Loss")
    plt.plot(loss, color="g", label="Train Loss")
    plt.ylabel("Loss")
    plt.xlabel("Epochs")
    plt.legend(loc="best")
    plt.show()
```

# Set Hyperparameters

```python
MARGIN = 0.501
FILTER_SIZES = [1, 2, 3, 4, 5]
NUM_FILTER = 128
DROPOUT_RATE = 0.5
BATCH_SIZE = 1000
NUM_EPOCHS = 100
STEPS_PER_EPOCH = 1
```

```python
gen_train = data_generator(xTrain, yTrain, BATCH_SIZE)
```

```python
mwcnn = MWCNN(margin=MARGIN, filter_sizes=FILTER_SIZES, num_filter=NUM_FILTER, dropout_rate=DROPOUT_RATE)
VAL_ACC = []
TEST_ACC = []
history = {}
for e in tqdm(range(NUM_EPOCHS)):
    #print('EPOCHS', e+1)
    t = mwcnn.fit_generator(gen_train,
                            steps_per_epoch=STEPS_PER_EPOCH,
                            epochs=1,
                            verbose=0,
                            workers=4,
                            use_multiprocessing=True)
    for i, j in t.history.items():
        history[i] = history.get(i, []) + j
    VAL_ACC.append(accuracy(mwcnn, x1Val, x2Val, yVal, show=False))
    TEST_ACC.append(accuracy(mwcnn, x1Test, x2Test, yTest, show=False))
    
print("Training Loss")
plot_loss(history)
print("Val Accuracy")
plot_acc(VAL_ACC, NUM_EPOCHS)
print("Test Accuracy")
plot_acc(TEST_ACC, NUM_EPOCHS)
```

```python
def softmax(x):
    """
    Compute the softmax function for each row of the input x.
    """
    orig_shape = x.shape
    if len(x.shape) > 1:
        tmp = np.max(x,axis=1)
        x-=tmp.reshape((x.shape[0],1))# here we use Numpy broadcasting
        x = np.exp(x)
        tmp = np.sum(x, axis = 1)
        x /= tmp.reshape((x.shape[0], 1))# here we use Numpy broadcasting
    else:
        # Vector 
        tmp = np.max(x)
        x -= tmp
        x = np.exp(x)
        tmp = np.sum(x)
        x /= tmp
    return x
```
