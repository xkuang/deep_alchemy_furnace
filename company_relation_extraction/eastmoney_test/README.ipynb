## 公司实体关系定义

公司实体间的关系可以用三元组（实体1，关系$r$，实体2）来描述，其中$\forall r_{1}\in R$，$\forall r_{2}\in R$，$r_{1}$与$r_{2}$之间不存在互相包含的情况（$R$为关系集合）。如在例子“A公司与B公司就某业务开展合作”中，关系可以被描述为（A，合作， B）。在东方财富网的公司新闻报道中有时会出现多公司实体的关系情况，例如“【东富龙：向迪安诊断转让重庆产研院股权】东富龙(300171)6月1日晚公告，公司下属全资子公司东富龙医疗与迪安诊断(300244)签署股权转让协议，约定东富龙医疗以5035万元，向迪安诊断转让重庆精准医疗产业技术研究院有限公司(简称重庆产研院)100%股权。（证券时报网）”。对于这种情况分析后，可以对其进行拆分使其符合三元组的描述方式，例句的拆分方法为：首先将新闻中公司实体按出现次序排序。
     
     本例中公司实体分别为：

然后将实体全排列两两组成一对，并按照以下标注规则，标注相对关系。

在东方财富网的公司新闻报道中，比较常见的实体关系有以下几种：

    （1）买卖。
    财经新闻中比较常见的商业行为，包括公司之间的收购、转让，公司股票之间的增持、减持、买进卖出等。需要注意的是公司之间买卖的媒介工具包括金钱及股票等，通过股票或股权进行的商业活动也被归为买卖行为。上例即为买卖关系的例子。
    （2）从属。
    即A公司是B公司的子公司等情况。在实际的商业活动中，很多行为都是由子公司出面进行的，如合作关系中所举之例：“万方发展1日晚间公告，公司参股子公司成都义幻医疗科技有限公司与腾讯云计算(北京)有限责任公司签订了《战略合作协议》”，由于“成都义幻医疗科技有限公司”是“万方发展”的参股子公司，便可推出“万方发展”和“腾讯云计算”是合作的关系。可见，从属关系对于更为深层次的关系研究有着重要的意义。
    （3）合并。
    合并和收购是不同的概念。后者是属于买卖大类的情况，合并则是指两个企业组成一个新的实体，原企业不以法律实体形式继续存在的情况。例如A公司与B公司合并成为C公司。合并包括吸收合并以及新设合并，吸收合并是指A公司将B公司吸收后，B公司解散，新设合并是指AB两家公司各方解散成立新公司。本定义中吸收合并与新设合并均计入合并大类。
    
    例：“【京山轻机8亿元并购案被问询】6月12日，深交所发布对京山轻机“发行股份及支付现金购买资产并募集配套资金暨关联交易报告书的事后审查问询函，涉及公司收购标的苏州晟成自设立至今包括出资情况、利润分配、产能、客户关系、研发和管理费用等在内的多个方面，并要求进行补充披露。（上海证券报）”
    
    上例中公司实体分别为：
    （京山轻机，苏州晟成）
    
    公司关系为：
    （京山轻机，合并，苏州晟成）-- 6
    （苏州晟成，合并，京山轻机）-- 6
    
    （4）合作。
    公司之间的合作是指公司间通过协议、签约或其他联合方式共同进行商业活动的一种关系。下例为合作关系的例子。
    
    例："子公司战略携手腾讯云计算， 使得万方发展在6月2日开盘后不久涨停，报11.68元。万方发展1日晚间公告，公司参股子公司成都义幻医疗科技有限公司与腾讯云计算(北京)有限责任公司签订了《战略合作协议》，就产业互联网、医疗云解决方案等相关领域展开深度合作。"
    
    上例中公司实体分别为：
    （腾讯云计算, 万方发展, 成都义幻医疗科技有限公司）
    
    公司关系为：
    （腾讯云计算，合作，万方发展）-- 7
    （腾讯云计算，合作，成都义幻医疗科技有限公司）-- 7
    （万方发展，合作，腾讯云计算）-- 7
    （万方发展，从属2，成都义幻医疗科技有限公司）-- 5
    （成都义幻医疗科技有限公司，合作，腾讯云计算）-- 7
    （成都义幻医疗科技有限公司，从属1，万方发展）-- 4
    
    （5）竞争。
    公司之间的竞争关系是指公司间采取力图胜过或压倒对方的商业行为活动而产生的关系。下例为竞争关系的例子。
    
    例：“顺丰控股今日早间发布公告，就投资者关注的“菜鸟争端”等问题作出说明。顺丰控股表示，预计此事不会影响去年重组时对2017年和2018年净利润做出的业绩承诺，并表示，事件发生后， 公司已通过媒体向公众及用户解释事件因由。据了解，顺丰6月1日凌晨宣布关闭对菜鸟的数据接口，中午已经停止给所有淘宝平台上的包裹回传物流信息。菜鸟方面表示，对此感到很突然，正在与顺丰紧急沟通。在此期间，为保证消费者和商家利益，菜鸟建议商家暂时改用其他快递公司发货。”
    
    上例中公司实体分别为：
    （顺丰控股，菜鸟）
    
    公司关系为：
    （顺丰控股，竞争，菜鸟）-- 8
    （菜鸟，竞争，顺丰控股）-- 8
    
    （6）

- 合并：组合、并购、并轨、吞并、归并、合并、融合、合股、合资、兼并

## 公司实体关系语料库标注规则

按照以上三元组（实体1，关系，实体2）定义，可以将标注内容分为2部分，分别为公司实体内容标注和公司关系内容标注。公司实体标注信息为公司名称，公司关系标注信息为以上定义的5种关系，分别为：买卖、从属、合并、合作、竞争。

财经新闻文本虽然用语规范，语法要求较高，但是在单个新闻中公司实体多，涉及关系复杂，指代现象普遍，且符号表示较多。遵照不同的标注方法与标注粒度，标注的关系结果也会不同。因此为保证语料库的规范性，在标注过程中需要遵守一定的标注规则，下面给出了一部分标注过程中需要遵守的标注规则：

    （1）在财经新闻中，常有某个公司全称和简称在同一新闻中共同出现的情况，在此情况下，视具体情况而定，一般优先使用简称。
    （2）在标注过程发现，部分公司实体对之间存在了不止一种关系。如：“亚洲资源(00899)发布公告，于2017年6月7日，该公司已完成收购汇联(中国)有限公司67%股权。完成后，目标公司将成为该公司的非全资附属公司。”，对于这种情况，应将实体对关系标注为时效性最强的关系或者新闻最想表达的关系，例如本例中公司实体（亚洲资源，汇联(中国)有限公司）间的关系应被标注为“从属”关系而非“买卖”关系。
    （3）在财经新闻中，同一实体会出现多次，在标注过程中忽略相同实体的位置顺序。
    （4）公司实体关系中除“从属”、“买卖”关系外不考虑实体位置，例如“（A，合作，B）”等同于“（B，合作，A）”。
    （5）对不属于五大类的公司实体关系，均归类为“其他”。
    （6）在标注公司实体关系时，以“1”表示关系“其他”，“2”表示关系“买卖（A为买方，B为卖方）”，“3”表示关系“买卖（A为卖方，B为买方）”，“4”表示关系“从属（A从属于B）”，“5”表示关系“从属（A下属B）”，“6”表示关系“合并”，“7”表示关系“合作”，“8”表示关系“竞争”。
    
根据以上规则对财经新闻语料进行标注，并将标注的结果储存在json文件中。一个完整的标注示例如下所示。

    {"abstract": "【东富龙：向迪安诊断转让重庆产研院股权】东富龙(300171)6月1日晚公告，公司下属全资子公司东富龙医疗与迪安诊断(300244)签署股权转让协议，约定东富龙医疗以5035万元，向迪安诊断转让重庆精准医疗产业技术研究院有限公司(简称重庆产研院)100%股权。（证券时报网）", "eneity": ["东富龙", "迪安诊断", "重庆产研院", "东富龙医疗"], "relation": "315252141431"}
    
其中“abstract”代表爬取的财经新闻摘要，”entity“代表公司实体，”relation“代表公司实体间的关系。
