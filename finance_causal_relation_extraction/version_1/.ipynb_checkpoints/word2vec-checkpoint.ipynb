```python
def load_data(path):
    """
    Load date from file
    """
    input_file = os.path.join(path)
    with open(input_file) as f:
        lines = f.readlines()
    return lines
```

```python
def load_mdata(path):
    """
    Load mission date from file
    """
    data = []
    input_file = os.path.join(path)
    with open(input_file) as f:
        lines = f.readlines()
    for line in lines:
        item = json.loads(line)
        data.append(item["contWords"])
    return data
```

```python
class data_for_word2vec:
    """
    Generate trianing data for word2vec
    """
    def __iter__(self):
        for c in content:
            item = json.loads(c)
            yield item['contWords']
```
